## Availability of marijuana as perceived by 12th graders in the U.S. from 1975 to 2021

```python
#listing files in our directory 

from os import listdir
listdir()
```

```python
#Import package
import pandas as pd

#Obtaining the data file we want
data_file = 'Availability_of_marijuana_1975_2021.csv'
df = pd.read_csv(data_file)

#Dataframe
df
```

```python
#Creating a list for your variables
Year = list(df['Year'])
availability_of_marijuana = list(df['availability_of_marijuana'])

print(Year,availability_of_marijuana)
```

```python
import matplotlib.pyplot as plt

#Input the yearly dates
x = 1975, 1980, 1985, 1990, 1995, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2021

#Sadly, python does not like percentages, so we got to take away the symbol
y = 88, 89, 86, 84, 89, 89, 89, 87, 87, 86, 86, 85, 84, 84, 81, 82, 82, 82, 81, 81, 80, 81, 80, 80, 78, 70

#Directions on how to set-up your figure

fig, ax = plt.subplots()

#We want a line plot
plt.plot(x, y)

#Labeling our title and axis 
plt.title('Fig.13: Availability of marijuana as perceived by 12th graders in the U.S. from 1975 to 2021')
plt.xlabel('Year')
plt.ylabel('Availability of marijuana (%)')

plt.show()
```
