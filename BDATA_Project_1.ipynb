# Relationship between Income and Needing But Not Receiving Treatment at a Specialty Facility for 12-17 year olds


```python
#Obtain our data file
import pandas as pd

data_file = 'income_treatment.csv'

df = pd.read_csv(data_file)

df
```

```python
#Making our data into lists
state = list(df['state'])
income = list(df['income'])
needing_but_not_receiving_treatment_alcohol_use = list(df['needing_but_not_receiving_treatment_alcohol_use'])
needing_but_not_receiving_treatment_substance_use = list(df['needing_but_not_receiving_treatment_substance_use'])

print(state,income,needing_but_not_receiving_treatment_alcohol_use,needing_but_not_receiving_treatment_substance_use)
```

```python
#Import packages
import matplotlib.pyplot as plt
import numpy as np

#List data for income
x = list(df['income'])

#List data for the population who needed but didn't receive treatment for alcohol use
y = list(df['needing_but_not_receiving_treatment_alcohol_use'])

#List data for the population needing but not receiving treatment for substance use
y2 = list(df['needing_but_not_receiving_treatment_substance_use'])

#Figure design

fig, ax = plt.subplots()
plt.scatter(x, y, c= 'maroon')
plt.scatter(x, y2, c= 'orange')

plt.title('Fig.10: U.S States Median Household Income and 12-17 year olds not receiving treatment for Alcohol/Substance Use')
plt.xlabel('Income')
plt.ylabel('Needing but not receiving treatment')

#Adding a legend, so reader know which color correspond to which variable
ax.plot(y, color='maroon', label='Alcohol')
ax.plot(y2, color='orange', label='Substance')

#Giving directions of the location of our legend
leg = ax.legend(loc='upper left')

plt.show()
```
